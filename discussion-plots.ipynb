```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
dont_panic=42


# Create a dataset with a normal distribution of about 1000 values
np.random.seed(dont_panic)
data = np.random.normal(loc=0, scale=1, size=100000)
print(data)
```

```python
sns.kdeplot(data, color='black')
plt.title('Two-sided Test')
plt.xlabel('')
plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)
plt.ylabel('')
mean = np.mean(data)
std_dev = np.std(data)
x = np.linspace(data.min(), data.max(), 1000)  
y = stats.norm.pdf(x, mean, std_dev)  

plt.fill_between(x, y, where=(x >= mean + 1.96 * std_dev), color='red') # Shade above
plt.fill_between(x, y, where=(x <= mean - 1.96 * std_dev), color='red') # Shade below

plt.axvline(mean + 1.96 * std_dev, color='r', linestyle='--', label='Critical Value')
plt.axvline(mean - 1.96 * std_dev, color='r', linestyle='--')

plt.text(mean, max(y)/2, '1-α', horizontalalignment='center', verticalalignment='center', fontsize=12, color='black')
plt.text(mean + 2.5 * std_dev, max(y)/4, 'α/2', horizontalalignment='center', verticalalignment='center', fontsize=12, color='black')
plt.text(mean - 2.5 * std_dev, max(y)/4, 'α/2', horizontalalignment='center', verticalalignment='center', fontsize=12, color='black')
plt.annotate('Critical Value', xy=(mean + 1.96 * std_dev, max(y)/2), xytext=(mean + 1.96 * std_dev + 0.2, max(y)/2),
             fontsize=12, color='red', rotation=90, verticalalignment='center')
plt.annotate('Critical Value', xy=(mean - 1.96 * std_dev, max(y)/2), xytext=(mean - 1.96 * std_dev - 0.5, max(y)/2),
             fontsize=12, color='red', rotation=90, verticalalignment='center')
plt.annotate('Reject H$_0$', xy=(mean + 1.96 * std_dev, max(y)/2), xytext=(mean + 2.5 * std_dev, max(y)/1.2), fontsize=12, color='black')
plt.annotate('Reject H$_0$', xy=(mean - 1.96 * std_dev, max(y)/2), xytext=(mean - 4.3 * std_dev, max(y)/1.2), fontsize=12, color='black')
plt.text(mean, max(y)/10, 'Do not reject H$_0$', fontsize=12, color='black', ha='center')

plt.show()
```

```python
sns.kdeplot(data, color='black')
plt.title('Right Tailed Test')
plt.xlabel('')
plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)
plt.ylabel('')
mean = np.mean(data)
std_dev = np.std(data)
x = np.linspace(data.min(), data.max(), 1000)  
y = stats.norm.pdf(x, mean, std_dev)  

plt.fill_between(x, y, where=(x >= mean + 1.65 * std_dev), color='red') # Shade above

plt.axvline(mean + 1.65 * std_dev, color='r', linestyle='--', label='Critical Value')

plt.text(mean, max(y)/2, '1-α', horizontalalignment='center', verticalalignment='center', fontsize=12, color='black')
plt.text(mean + 2.5 * std_dev, max(y)/4, 'α', horizontalalignment='center', verticalalignment='center', fontsize=12, color='black')

plt.annotate('Critical Value', xy=(mean + 1.65 * std_dev, max(y)/2), xytext=(mean + 1.96 * std_dev + 0.2, max(y)/2),
             fontsize=12, color='red', rotation=90, verticalalignment='center')

plt.annotate('Reject H$_0$', xy=(mean + 1.96 * std_dev, max(y)/2), xytext=(mean + 2.5 * std_dev, max(y)/1.2), fontsize=12, color='black')

plt.text(mean, max(y)/10, 'Do not reject H$_0$', fontsize=12, color='black', ha='center', va='bottom')

plt.show()
```
